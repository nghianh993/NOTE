# Lấy dữ liệu và training sau đó upload lên google drive

```python
import requests
import numpy as np
from datetime import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from google.colab import drive
import pickle

drive.mount('/content/drive')
url = "https://rongbachkim.com"
now = datetime.now()

# Gửi yêu cầu HTTP đến API và nhận dữ liệu JSON
response = requests.get(url + '/ketqua.php?getkq&ngay=2023-12-07&days=300&wday=0')
data = response.json()

# Chuyển đổi dữ liệu thành đặc trưng (features) và kết quả (labels)
features = []
labels = []
for day in data:
    date = day[0].replace('-', '')
    results = day[1]
    for result in results:
        features.append([float(result)])
        labels.append([float(date)])

features = np.array(features[::-1], dtype=np.float32)
labels = np.array(labels[::-1], dtype=np.float32)

# Tạo mô hình LSTM
model = Sequential()
model.add(LSTM(64, input_shape=(1, 1)))
model.add(Dense(1))

# Biên dịch mô hình
model.compile(optimizer='adam', loss='mse')

# Huấn luyện mô hình
model.fit(features, labels, epochs=1000, batch_size=64)

with open('/content/drive/MyDrive/Model20231207.pkl', 'wb') as file:
    pickle.dump(model, file)

X_next_day = np.array([features[-1]])
y_predicted = model.predict(X_next_day)

# Chuyển đổi kết quả dự đoán thành kiểu nguyên
y_predicted = y_predicted.flatten().astype(int)

# Tạo kết quả dưới dạng mảng tương tự như dữ liệu đã cho
now = datetime.now()
next_date = now.strftime("%Y-%m-%d")
next_results = [str(num).zfill(5) for num in y_predicted]
next_data = [next_date, next_results]

# In kết quả dự đoán
print(next_data)
```

# Lấy Model đã xây dựng và tiếp tục training với kết quả 30 ngày 1000 lần

```python
import requests
import numpy as np
from datetime import datetime
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from google.colab import drive
import pickle

drive.mount('/content/drive')
model_path = '/content/drive/MyDrive/AIKQXS.pkl'
url = "https://rongbachkim.com"
now = datetime.now()
next_date = now.strftime("%Y-%m-%d")
dateAI = next_date.replace('-', '')
# Gửi yêu cầu HTTP đến API và nhận dữ liệu JSON
response = requests.get(url + '/ketqua.php?getkq&ngay=' + next_date + '&days=30&wday=0')
data = response.json()

# Chuyển đổi dữ liệu thành đặc trưng (features) và kết quả (labels)
features = []
labels = []
for day in data:
    date = day[0].replace('-', '')
    results = day[1]
    for result in results:
        features.append([float(result)])
        labels.append([float(date)])

features = np.array(features[::-1], dtype=np.float32)
labels = np.array(labels[::-1], dtype=np.float32)

results = []

for number in range(1, 7):
    with open(model_path, 'rb') as file:
        model = pickle.load(file)

    # Huấn luyện mô hình
    model.fit(features, labels, epochs=1000, batch_size=27)

    #Lưu lại model đã huấn luyện
    with open(model_path, 'wb') as file:
        pickle.dump(model, file)

    X_next_day = np.array([features[-1]])
    y_predicted = model.predict(X_next_day)

    # Chuyển đổi kết quả dự đoán thành kiểu nguyên
    y_predicted = y_predicted.flatten().astype(int)

    # Tạo kết quả dưới dạng mảng tương tự như dữ liệu đã cho
    next_results = [str(num).zfill(5) for num in y_predicted]
    results = results + next_results;

# In kết quả dự đoán
print([next_date, results])
```
